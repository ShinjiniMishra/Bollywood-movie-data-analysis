```python
import pandas as pd
df = pd.read_csv('Bollywood movies dataset.csv')
df.head(20)
```

#Data cleaning
Removing special characters and checking for missing values


df = df.apply(lambda x: x.str.replace('?', 'a', regex=False) if x.dtype == 'object' else x)

#Data visualization
1. Plotting average ratings Per genre in bollywood
2. Plotting average bollywood movie rating over the years
3. Plotting of number of movies released per year
4. Plotting number of movies per genre
5. Plotting of a correlation heat map depicting correlation between different columns of the data set

```python
import matplotlib.pyplot as plt

genre_ratings = df.groupby('Genre')['IMDb Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.bar(genre_ratings.index, genre_ratings.values, color='brown')
plt.xticks(rotation=45, ha='right')
plt.title('Average IMDb Rating by Genre in Bollywood')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
year_ratings = df.groupby('Year')['IMDb Rating'].mean()

plt.figure(figsize=(12,6))
plt.plot(year_ratings.index, year_ratings.values, color='steelblue', linewidth=2)
plt.title('Average Bollywood Movie Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
movies_per_year = df.groupby('Year')['Movie Title'].count()

plt.figure(figsize=(12,6))
plt.bar(movies_per_year.index, movies_per_year.values, color='mediumpurple')
plt.title('Number of Bollywood Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(movies_per_year.index, rotation=90, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
movies_per_genre = df.groupby('Genre')['Movie Title'].count().sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.bar(movies_per_genre.index, movies_per_genre.values, color='steelblue')
plt.title('Number of Movies per Genre in Bollywood')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5)
plt.title('Correlation Heatmap of Bollywood Movies')
plt.tight_layout()
plt.show()
```

The following are the findings from my data analysis:

1.Genre    
Genre that dominated mostly throughout 2017 to 2025 was War/Biographical and thus had higher average ratings in the dataset

Genre with the least average rating was Action/Fantasy in the dataset

2.Year   
2021 was the year with the highest rated movies in the dataset

2019 and 2023 had movies with the lowest ratings

2017, 2024 and 2025 had the maximum number of movie releases in the dataset

2020 and 2021 had the least number of movie releases mainly due to the Covid pandemic

3.Correlation Heatmap Findings   
Budget of a movie and its rating have a negative correlation, meaning the rating of a movie is not influenced by its budget. A big budget movie does not ensure high ratings

Actor count and budget have a strong positive correlation, meaning more the actor count, higher will be the budget of that particular movie

Revenue and profit also have a strong positive correlation, meaning higher the revenue, higher will be the profit obtained from that movie

Budget and profit of a movie have no correlation with the year in which it is released. This is expected as production costs and market sizes differ significantly across years

Movies with higher or lower profit do not necessarily have high or low ratings. Rating of a movie has no correlation with its profit

Actor count and revenue of a movie have no correlation, signifying that a movie with more or less actors does not guarantee high or low revenue
